```python
import pandas as pd
import os
import csv
from os.path import join
import numpy as np
colnames = ['Subject_id','ID','Left_Amygdala','Right_Amygdala','Diagnosis'] 
df = pd.read_csv('/cis/home/sphatak/hello_world_project/data/measures/Ref_measures_amygdala.csv',names = colnames)
id_list = df.ID.tolist()
del id_list[0]
for i in range(0, len(id_list)): 
    id_list[i] = int(id_list[i]) 
diagnosis_list = df.Diagnosis.tolist()
del diagnosis_list[0]
print(len(id_list))
print(len(diagnosis_list))
x_train = []
normal = ['NCF', 'NCO'] 
path = "/cis/home/sphatak/hello_world_project/data/CSVDataFiles"
os.chdir(path)
for filename in os.listdir(path):
    if filename.endswith(".csv"):
        subject_id = int(filename[12:19])
        sub_idx = id_list.index(subject_id)
        label = diagnosis_list[sub_idx]
        boo = label not in normal
        features_df = pd.read_csv(filename, header=None)
        feature_list = features_df.ix[:,1]
        feature_list = feature_list.tolist()
        if(boo):
            feature_list.append(1)
        else:
            feature_list.append(0)
        x_train.append(feature_list)
        
print (x_train)
print (len(x_train))
print (len(x_train[0]))
f = np.array(x_train)
pd.DataFrame(x_train).to_csv("/cis/home/sphatak/hello_world_project/data/features.csv")
```
